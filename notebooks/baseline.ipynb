```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
DATA_DIR = '../data'
X_TRAIN_PATH = os.path.join(DATA_DIR, 'X_train.csv')
Y_TRAIN_PATH = os.path.join(DATA_DIR, 'y_train.csv')
X_TEST_PATH = os.path.join(DATA_DIR, 'X_test.csv')

X_train = pd.read_csv(X_TRAIN_PATH, index_col='ID')
y_train = pd.read_csv(Y_TRAIN_PATH, index_col='ID')
X_test = pd.read_csv(X_TEST_PATH, index_col='ID')

cols_to_drop = ['DATE', 'STOCK', 'INDUSTRY', 'INDUSTRY_GROUP', 'SUB_INDUSTRY', 'SECTOR']
X_train_clean = X_train.drop(cols_to_drop, axis=1, errors='ignore').fillna(0)
X_test_clean = X_test.drop(cols_to_drop, axis=1, errors='ignore').fillna(0)

X_tr, X_val, y_tr, y_val = train_test_split(X_train_clean, y_train['RET'], test_size=0.2, random_state=42)

model_val = RandomForestClassifier(n_estimators=100, max_depth=8, random_state=42, n_jobs=-1)
model_val.fit(X_tr, y_tr)
val_acc = accuracy_score(y_val, model_val.predict(X_val))

model_final = RandomForestClassifier(n_estimators=100, max_depth=8, random_state=42, n_jobs=-1)
model_final.fit(X_train_clean, y_train['RET'])

submission_preds = model_final.predict(X_test_clean)
submission = pd.DataFrame({'RET': submission_preds}, index=X_test_clean.index)
submission.to_csv('submission_baseline.csv')

print(f"Execution complete.\nInternal Validation Accuracy: {val_acc:.4f}\nSubmission file 'submission_baseline.csv' has been generated.")
```
